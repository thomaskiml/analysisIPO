```python
import os

luto = []


def timePeriod(i):
    for file in os.listdir("./marketData"):
        try:
            unpickled_df = pd.read_pickle("./marketData/"+file).iloc[:i]
            ticker1 = file.replace('.pkl', '')
        
            min = float(unpickled_df["Low"].min())
            minDate = str(unpickled_df['Low'].idxmin())[:10]
            max = float(unpickled_df['High'].max())
            change = (max/min)*100
            maxDate = str(unpickled_df['High'].idxmax())[:10]
            avgVol = int(unpickled_df["Volume"].mean())
            avgClo = float(unpickled_df["Close"].mean())
            data = {'Ticker':ticker1, 'Min':min, 'Min Date':minDate, 'Max':max, 'Max Date':maxDate, 'Change %':change, 'Average Close':avgClo, 'Average Volume':avgVol}
            
            luto.append(data)
        except:
            pass

for i in (7, 14, 30, 90, 365):
    luto.clear()
    timePeriod(i)
    frame = pd.DataFrame(luto)
    frame.to_pickle("./sets/"+str(i)+"Day.pkl")

```
